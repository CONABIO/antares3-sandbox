# Comparison

```python
for k in range(len(lista_iter_10m_sorted)):
    tile_10m = lista_iter_10m_sorted[k]
    tile_20m = lista_iter_20m_sorted[k]
    dc_tile_10m = tile_10m[0]
    dc_tile_20m = tile_20m[0]
    number_of_images_10m = tile_10m[1].shape[0]
    number_of_images_20m = tile_20m[1].shape[0] 
    if number_of_images_10m != number_of_images_20m:
        ids_10m = [str(j.values.item()[0].id) for j in tile_10m[1].sources]
        ids_20m = [str(j.values.item()[0].id) for j in tile_20m[1].sources]
        paths_10m = [str(j.values.item()[0].local_path) for j in tile_10m[1].sources]
        paths_20m = [str(j.values.item()[0].local_path) for j in tile_20m[1].sources]
        print(''.join(["dc tile 10m: ",
                       str(dc_tile_10m),
                       "dc tile 20m: ",
                       str(dc_tile_20m)
                      ])
             )
        
        print(''.join(["number of images 10m:", 
                       str(number_of_images_10m), " ",
                       "number of images 20m:",
                       str(number_of_images_20m)
                       ])
             )
        print("-"*20)
        print("ids of 10m:")
        print(ids_10m)
        print("-"*20)
        print("length paths of 10m:")
        print(len(paths_10m))
        print("-"*20)
        print("paths of 10m:")
        print(paths_10m)
        print("-"*20)
        print("length paths of 20m:")
        print(len(paths_20m))
        print("-"*20)
        print("paths of 20m:")
        print(paths_20m)
        print("-"*20)
        print("ids of 20m:")
        print(ids_20m)
        print("-"*20)
```
