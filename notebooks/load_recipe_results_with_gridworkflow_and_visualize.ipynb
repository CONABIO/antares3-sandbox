```python
import datacube
%matplotlib inline
import matplotlib
from datacube.api import GridWorkflow
from madmex.wrappers import gwf_query
```

```python
name_of_product = 'landsat_belize_001_recipe_2015'
gwf_kwargs = {'region': 'BLZ', 'begin': '2013-01-01', 'end':'2015-12-31'}
gwf_kwargs.update(product=name_of_product)
```

```python
iterable = gwf_query(**gwf_kwargs)
```

```python
lista_iter = list(iterable)
var = lista_iter[1]
```

```python
sr_0 = GridWorkflow.load(var[1],dask_chunks={'longitude': 2000, 'latitude': 2000, 'time': 1})
```
